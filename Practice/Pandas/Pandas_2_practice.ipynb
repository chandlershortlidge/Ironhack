# Concat
- stacks by index
- be careful about index, it will compare indexes

## Split type #1: data is split by index, eg, both tables have the same features but different data samples 

```python
df1 = df[df["Department"] == "Sales"]
df2 = df[df["Department"] == "Research & Development"]
```

```python
pd.concat(
    [df1, df2]
)
```

## data split type 2 — data is split along features — both dataframes have the data samples, but different features
- we want to stack tables next to each other 
- do it using pd.concat and axis=1
- axis=1 means "stack horizontally"

```python
## data split type 2 — data is split along features — both dataframes have the data samples, but different features

sd1 = df[['Department', 'JobRole', 'MonthlyIncome', 'EmployeeNumber']] # four columns
sd2 = df[['BusinessTravel', 'EducationField', 'EmployeeNumber', 'EnvironmentSatisfaction']] # four columns
```

```python
pd.concat(
    [sd1, sd2], axis=1 # eight columns
)
```

```python
pd.concat(
    [df1, df2],
    axis=1
)
```

- axis=0 
- stack vertically 
- stack by rows 

```python
pd.concat(
    [sd1, sd2], axis=0 
)
```

# Merging and Joining

```python
df1 = pd.DataFrame(
    {
        "employee_id": ["E0", "E0", "E2", "E3"],
        "department": ["D0", "D1", "D2", "D1"],
        "salary": ["S0", "S1", "S2", "S3"],
        "bonus": ["B0", "B1", "B2", "B3"],
    }
)

df2 = pd.DataFrame(
    {
        "employee_id": ["E0", "E1", "E2", "E3"],
        "department": ["D0", "D0", "D1", "D1"],
        "JobRole": ["JR0", "JR1", "JR2", "JR3"],
        "JobLevel": ["JL0", "JL1", "JL2", "JL3"],
    }
)

df3 = pd.DataFrame(
    {
        "employee_id": ["E90", "E1", "E2", "E3"],
        "department": ["D0", "D1", "D2", "D2"],
        "EducationField": ["EF0", "EF1", "EF2", "EF3"],
        "EnvironmentSatisfaction": ["ES0", "ES1", "ES2", "ES3"],
    }
)
```

*Merge*

```python
from IPython.display import display_html 
# Just showing tables
df1_styler = df1.style.set_table_attributes("style='display:inline'").set_caption('Left table')
df2_styler = df2.style.set_table_attributes("style='display:inline'").set_caption('Right table')
display_html(df1_styler._repr_html_() + " " + df2_styler._repr_html_(), raw=True)
```

```python
pd.merge(df1, df2, how="inner")
# first pandas finds common columns 
# then it only uses data rows that have identical values in these common columns in both tables
# intersections or AND opperation
```

```python
pd.merge(df1, df2, how="left") #merges left table & right table

# finds common coluns --> emplyee id and department
# takes all of table 1
# searchs for matches in table 2
# if they match, they are included

```

```python
pd.merge(df1, df2, how="right") #merges right table & left table
```

```python
pd.merge(df1, df2, how="outer")
```

```python
pd.merge(df1, df2, how="cross") # all possible combinations of rows in both tables
#table 1 -> n rows 
# table 2 -> n rows 
# output will be n*n
```
