#### Obtain the distinct values across `Type 1` and `Type 2`.

Exctract all the values in `Type 1` and `Type 2`. Then create an array containing the distinct values across both fields.

```python
df["Type 1"] # df["Type 1"] gives you a pandas Series with values and index
```

```python
print(type(df["Type 1"])) # Series type
print(type(df["Type 1"].to_list())) # list type
```

```python
type1 = df["Type 1"].to_list()
type2 = df["Type 2"].to_list() # to_list() converts the Series int into a list in order to combine them using concat

type2
```

```python
combined = type1 + type2 # concat the lists
```

### set()
- A set in Python only keeps unique values—it removes any duplicates.
- **Sets are unordered and don’t support indexing** 

```python
distinct = set(combined) # turn combined into a set and assign to new variable
```

### List 
- Lists support indexing, slicing, and other operations that are often useful in data analysis and plotting.
- Some functions or libraries (like pandas, NumPy, plotting tools) expect lists or arrays—not sets.

```python
distinct_list = list(distinct) # turn set back into a list
distinct_list 
```

### Creating an array with np.array

```python
distinct_array = np.array(distinct_list) # create a new varaible to assign our array to
print(distinct_array)
```

#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
name_list = list(df.Name)
name_list # series
```

```python
def fix_mega(name):
    # If "Mega" is in the name, move it to the front
    if "Mega" in name:
        idx = name.find("Mega")
        return "Mega" + name[idx + 4:]  # Get everything after "Mega"
    else:
        return name

# Now apply this function to each name in your list using .apply()
df["Name"] = df["Name"].apply(fix_mega)

df
```

#### Create a new column called `A/D Ratio` whose value equals to `Attack` devided by `Defense`.

For instance, if a pokemon has the Attack score 49 and Defense score 49, the corresponding `A/D Ratio` is 49/49=1.

```python
df["A/D Ratio"] = df["Attack"] / df["Defense"] 

print(df)

# df["new_col"] = df["col1"] / df["col2"]
```

#### Identify the pokemon with the highest `A/D Ratio`.

```python
df.loc[df["A/D Ratio"].idxmax()] 
```

#### Identify the pokemon with the lowest A/D Ratio.

```python
df.loc[df["A/D Ratio"].idxmin()]
```

#### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
df["Combo Type"] = df["Type 1"] + df["Type 2"].fillna("") 

df
```

#### Identify the pokemon whose `A/D Ratio` are among the top 5.

```python
top5 = df.nlargest(5, "A/D Ratio")
print(top5)
```

```python
last5 = df.nsmallest(5, "A/D Ratio")
print(last5)
```

#### For the 5 pokemon printed above, aggregate `Combo Type` and use a list to store the unique values.

Your end product is a list containing the distinct `Combo Type` values of the 5 pokemon with the highest `A/D Ratio`.

```python
combo_list = top5["Combo Type"].unique().tolist() 
#.unique() gives you the distinct values in the Series.
# .tolist() turns it into a list.
print(combo_list)
```
