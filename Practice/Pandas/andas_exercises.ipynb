### Use pd.to_numeric() to check for numeric values.
- This function tries to convert values to numbers. 
- If it can’t, it will set them as NaN (Not a Number) if you use errors='coerce'.
- This will give you a Series where non-numeric values become NaN.

```python
pd.to_numeric(result["hv1_y"], errors='coerce')
```

### To find out which rows are NOT numbers:
- isna()

```python
mask = pd.to_numeric(result["hv1_y"], errors='coerce').isna()
print(result[mask])
```

### us .loc to locate...

```python
print(result.loc[mask, "hv1_y"])
```

- Convert the column to numbers, turning bad values into NaN (missing data).
- That way, you can decide later if you want to drop those rows or fill them.
- This makes a new column (hv1_y_clean) that is all numbers, but will be NaN wherever the data wasn’t numeric.



```python
result["hv1_y_clean"] = pd.to_numeric(result["hv1_y"], errors='coerce')

print(result[["hv1_y", "hv1_y_clean"]].head(10))
```

```python
print(result.loc[mask, "hv1_y_clean"])
```

### Option 1: Drop NaNs first, then convert to int

```python
cleaned = result.dropna(subset=["hv1_y_clean"])
cleaned["hv1_y_clean"] = cleaned["hv1_y_clean"].astype(int)
```
