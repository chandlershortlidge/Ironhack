### compile together


    .: Matches any character except \n
    \.: seaches for literal "."
    *: Zero or more occurrences
    +: One or more occurrences
    ?: Zero or one occurrence
    [ ]: A set of characters
    { }: A specific number of occurrences
    \w: any letter (a-z, A-Z, 0-9, _)
    \b: end of the word
    {,3} 0 to 3 repetitions
    {3,} 3 or more repetitions
    {3,6} 3-6 repetitions
    \d: any digit (0-9)
    \D: any non-digit
    \s: any whitespace character (space, tab, newline)
    \S: any non-whitespace character

**Core functions:**

- `search()` - finds the first match in a string, returns a Match object or None
- `findall()` - finds all non-overlapping matches, returns a list of strings (or tuples if groups)
- `match()` - checks if pattern matches at the start of the string only
- `finditer()` - like findall but returns an iterator of Match objects
- `fullmatch()` - checks if the entire string matches the pattern

**Modification functions:**

- `sub()` - replaces matches with a replacement string
- `subn()` - like sub but also returns the number of substitutions made
- `split()` - splits string at pattern matches
